# Successive Halving
Successive Halving [2], originally Sequential Halving [2], is a multi-fidelity hyperparameter optimization technique.
It is explained in [2] as: 

> Given an input budget, uniformly allocate the budget to a set of arms \[hyperparameter configurations\] for a predefined amount of iterations \[e.g., epochs\], evaluate their performance, throw out the worst half, and repeat until just one arm remains.

[1] Karnin, Zohar, Tomer Koren, and Oren Somekh. "Almost optimal exploration in multi-armed bandits." International Conference on Machine Learning. PMLR, 2013.

[2] Jamieson, Kevin, and Ameet Talwalkar. "Non-stochastic best arm identification and hyperparameter optimization." Artificial Intelligence and Statistics. PMLR, 2016.

```python
import seaborn
import matplotlib.pyplot as plt
```

```python
configurations = [
    [(1, 0.45)],
    [(1, 0.53)],
    [(1, 0.55)],
    [(1, 0.50)],
    [(1, 0.60), (3, 0.75)],
    [(1, 0.63), (3, 0.7)],
    [(1, 0.65), (3, 0.8), (7, 0.85)],
    [(1, 0.70), (3, 0.77),(7, 0.87), (15, 0.88)],
]

fig, ax = plt.subplots(figsize=(8, 4))

for c in configurations:
    ax.scatter(*zip(*c))
    ax.plot(*zip(*c))
    
ax.set_title("Successive Halving\n (n=8, B=8, $\eta$=2)", fontsize=32)
ax.set_ylabel("Performance", fontsize=26)
ax.set_xlabel("Iterations trained", fontsize=26)
ax.tick_params(labelsize=22)
ax.set_xticks([1, 3, 7, 15])
```

```python
from matplotlib.backends.backend_pdf import PdfPages

with PdfPages("SH-iterations.pdf") as f:
    f.savefig(fig, bbox_inches="tight")
```
